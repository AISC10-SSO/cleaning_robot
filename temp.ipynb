```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# np.random.seed(42)
N = 100
X = []
for _ in range(N):
    x = np.random.randn(2) + np.array([1, 2]).T
    X.append(x)

X = np.stack(X, axis=0).T
plt.plot(X[0, :], X[1, :], '.')
```

```python
A = np.array([[0.5, 0], [0, 1]])
X1 = A @ X
plt.plot(X1[0, :], X1[1, :], '.')
```

```python
B = np.array([[1, 0], [0, -1]])
X2 = B @ X1
plt.plot(X2[0, :], X2[1, :], '.')
```

```python
C = np.array([[0, 1], [1, 0]])
X3 = C @ X2
plt.plot(X3[0, :], X3[1, :], '.')
```

```python
D = np.array([[np.cos(45), np.sin(45)], [-np.cos(45), np.sin(45)]])
X4 = D @ X3
plt.plot(X4[0, :], X4[1, :], '.')
```

```python
E = D @ C @ B @ A
X5 = E @ X
plt.plot(X5[0, :], X5[1, :], '.')
```

```python
F = A @ B @ C @ D
X6 = F @ X
plt.plot(X6[0, :], X6[1, :], '.')
```
